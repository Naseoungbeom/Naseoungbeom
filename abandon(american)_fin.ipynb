```python
import math #abandon

print("="*14,'Abandon American','='*14)

S = float(input('S : '))
X = float(input('X : '))
T = int(input('T : '))
V = float(input('V : '))
r = float(input('r : '))
T_interval = int(input('T_interval : '))

print('='*50)

u = math.exp(V * (T_interval**0.5))  
d = 1 / u  
p = (math.exp(r * T_interval) - d) / (u - d) 

results = []

print('<forward>')

for i in range(T + 1):  
    year_result = [] 

    if i == 0:
        year_result.append(round(S, 2)) 

    else:
        for j in range(i + 1):  
            value = S * (u ** (i - j)) * (d ** j) 
            year_result.append(round(value, 2)) 

    results.append(year_result)

    print(f"{i}year: {', '.join(map(str, year_result))}")

print('='*50)
fix_result = []
for i in range(len(results[T])):
    if results[T][i] > X:
        fix_result.append(results[T][i])
    else:
        fix_result.append(X)
fix_result = [round(x,2) for x in fix_result]


print('**abandon option apply for last list**')
print(fix_result)
results[T] = fix_result
print(results)
backward_results = [None] * (T + 1) 
print('='*50)
print('<back>')

backward_results[0] = results[T]
print(f"n year: {', '.join(map(str, results[T]))}")


for i in range(T - 1, -1, -1):  
    next_year_result = []

    for j in range(len(results[i])):
        value = ((p * backward_results[T - i - 1][j]) +
                 (1 - p) * backward_results[T - i - 1][min(j + 1, len(backward_results[T - i - 1]) - 1)]) \
                 * math.exp(-r * T_interval)
        value = round(value, 2)

        if value < X:
            value = round(X, 2)
        else:
            value = round(value,2)
            
        next_year_result.append(value)


    backward_results[T - i] = next_year_result


    print(f"{i}year: {', '.join(map(str, next_year_result))}")
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import math

# 파라미터 설정
V = 0.35
r = 0.05
T_interval = 1
u = math.exp(V * (T_interval**0.5))  # 상승 비율
d = 1 / u  # 하락 비율
p1 = (math.exp(r * T_interval) - d) / (u - d)  # 위험 중립 확률

def plot_binomial_lattice(forward_data, backward_data, p=round(p1, 2), save_path="expand.png"):
    G = nx.Graph()  # 방향 없는 그래프 사용
    pos = {}
    steps = len(forward_data) - 1  # 단계 수 계산
    
    # 노드와 엣지 생성
    for i in range(steps + 1):
        for j in range(i + 1):
            # 전방 계산 (주가)와 후방 계산 (옵션 가치) 값을 함께 노드에 표시
            forward_price = forward_data[i][j]
            backward_value = backward_data[i][j]
            G.add_node((i, j), forward_price=forward_price, backward_value=backward_value)
            pos[(i, j)] = (i, -j)  # X좌표는 i, Y좌표는 -j로 배치
            
            # 양방향 엣지 추가
            if i < steps:
                G.add_edge((i, j), (i + 1, j), weight=round(1-p, 2))
                G.add_edge((i, j), (i + 1, j + 1), weight=round(p, 2))

    # 그래프 시각화 (가로 길이 확대 및 폰트 크기 조정)
    plt.figure(figsize=(16, 8))
    nx.draw(G, pos, with_labels=False, node_size=1500, node_color="lightblue", node_shape="o", font_size=8)
    
    # 각 노드에 전방/후방 계산 값 표시 (소수점 제거 및 작은 폰트 크기)
    for node, attr in G.nodes(data=True):
        x, y = pos[node]
        plt.text(x, y, f"{int(attr['forward_price'])}\n{int(attr['backward_value'])}", 
                 ha='center', va='center', color="black", fontsize=8)

    # 각 엣지에 p와 1-p 확률 표시
    edge_labels = {(i, j): f"{G[i][j]['weight']:.2f}" for i, j in G.edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red", font_size=7)
    
    # 상단에 T=1, T=2, ... 표시
    for i in range(steps + 1):
        plt.text(i, 1, f"T={i}", ha='center', va='center', fontsize=10, color="black")

    # 축 설정 및 제목을 아래로 이동
    plt.xlabel("Steps (i)")
    plt.ylabel("Nodes (j)")

    # 고해상도로 이미지 저장
    plt.savefig(save_path, dpi=300, bbox_inches='tight')  # DPI를 높여 해상도 설정
    plt.show()

# 실행
plot_binomial_lattice(forward_data, backward_data)
```
